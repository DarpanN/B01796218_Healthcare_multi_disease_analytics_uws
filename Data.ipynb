```python
#Data Generation 
import numpy as np
import pandas as pd
import random
from datetime import datetime, timedelta

np.random.seed(42)
random.seed(42)

# -------------------------------
# Define constants
# -------------------------------
N = 100000
diseases = ['Heart Disease', 'Breast Cancer', 'Lung Cancer', 'Diabetes']
genders = ['Male', 'Female']
ethnicities = ['Caucasian', 'African American', 'Asian', 'Hispanic', 'Other']
socioeconomic_index = ['Low', 'Medium', 'High']
insurance_types = ['Private', 'Public', 'None']

# -------------------------------
# Initialize DataFrame
# -------------------------------
data = pd.DataFrame()

# -------------------------------
# Basic patient info
# -------------------------------
data['Patient_ID'] = [f'P{str(i).zfill(6)}' for i in range(1, N+1)]
data['Age'] = np.random.randint(0, 131, size=N)
data['Gender'] = np.random.choice(genders, size=N)
data['Ethnicity'] = np.random.choice(ethnicities, size=N)
data['Socioeconomic_Index'] = np.random.choice(socioeconomic_index, size=N)
data['Insurance_Type'] = np.random.choice(insurance_types, size=N)
data['Distance_to_Clinic_km'] = np.round(np.random.uniform(0.5, 50, N), 2)
data['Education_Level'] = np.random.choice(['None', 'High School', 'Bachelor', 'Master', 'PhD'], size=N)

# -------------------------------
# Disease assignment (balanced)
# -------------------------------
data['Primary_Diagnosis'] = np.random.choice(diseases, size=N)
data['Mortality_Event'] = np.random.choice([0,1], size=N, p=[0.8,0.2])
data['Vital_Status'] = np.where(data['Mortality_Event']==1, 'Deceased', 'Alive')
data['Comorbidity_Index'] = np.random.randint(0, 10, size=N)

# -------------------------------
# Temporal features
# -------------------------------
current_year = datetime.now().year
data['Diagnosis_Year'] = np.random.randint(current_year-50, current_year+1, size=N)
data['Diagnosis_Month'] = np.random.randint(1, 13, size=N)
data['Visit_Year'] = np.random.randint(data['Diagnosis_Year'], current_year+1)
data['Visit_Month'] = np.random.randint(1, 13, size=N)

# -------------------------------
# Lifestyle & lab features
# -------------------------------
data['BMI'] = np.round(np.random.normal(27, 5, N), 1)
data['Daily_Steps_Avg'] = np.random.randint(1000, 20000, size=N)
data['Sleep_Quality_Score'] = np.round(np.random.uniform(0, 10, N),1)
data['Stress_Cortisol_Level'] = np.round(np.random.uniform(5, 30, N),2)

# -------------------------------
# Cardiac biomarkers
# -------------------------------
data['NT_proBNP_pg_mL'] = np.round(np.random.normal(150, 50, N),1)
data['Troponin_T_ng_mL'] = np.round(np.random.normal(0.02, 0.01, N),3)
data['CRP_mg_L'] = np.round(np.random.normal(2, 1, N),2)
data['Ejection_Fraction_Pct'] = np.round(np.random.uniform(30, 70, N),1)
data['Systolic_BP'] = np.random.randint(90, 180, N)
data['Diastolic_BP'] = np.random.randint(60, 120, N)
data['Resting_HR'] = np.random.randint(50, 110, N)

# -------------------------------
# Oncology biomarkers
# -------------------------------
data['CEA_Level'] = np.round(np.random.uniform(0, 10, N),2)
data['CYFRA_21_1'] = np.round(np.random.uniform(0, 5, N),2)
data['EGFR_Mutation'] = np.random.choice([0,1], size=N)
data['ALK_Rearrangement'] = np.random.choice([0,1], size=N)
data['PDL1_Expression_Level'] = np.random.uniform(0, 100, N)
data['Histological_Type'] = np.random.choice(['Type A','Type B','Type C'], size=N)
data['TNM_Stage'] = np.random.choice(['I','II','III','IV'], size=N)
data['BRCA_Mutation'] = np.random.choice([0,1], size=N)
data['TP53_Mutation'] = np.random.choice([0,1], size=N)
data['PIK3CA_Mutation'] = np.random.choice([0,1], size=N)
data['Tumor_Size_mm'] = np.round(np.random.uniform(5, 100, N),1)
data['Lymph_Nodes_Involved'] = np.random.randint(0,10,N)
data['Tumor_Grade'] = np.random.choice([1,2,3], size=N)

# -------------------------------
# Metabolic / renal labs
# -------------------------------
data['HbA1c'] = np.round(np.random.uniform(4, 14, N),1)
data['Fasting_Blood_Glucose'] = np.round(np.random.uniform(70, 250, N),1)
data['LDL_Cholesterol'] = np.round(np.random.uniform(50, 200, N),1)
data['HDL_Cholesterol'] = np.round(np.random.uniform(20, 80, N),1)
data['Triglycerides'] = np.round(np.random.uniform(50, 300, N),1)
data['Urinary_ACR'] = np.round(np.random.uniform(0, 300, N),2)
data['eGFR'] = np.round(np.random.uniform(30, 120, N),1)

# -------------------------------
# Other features
# -------------------------------
data['FEV1_FVC_Ratio'] = np.round(np.random.uniform(0.5, 1.0, N),2)
data['Pack_Years'] = np.random.randint(0, 60, N)
data['Smoking_Status'] = np.random.choice(['Never','Former','Current'], size=N)
data['Occupational_Exposure'] = np.random.choice(['Low','Medium','High'], size=N)
data['Air_Quality_Index_Residence'] = np.round(np.random.uniform(0,500, N),1)
data['Family_History_Type2'] = np.random.choice([0,1], size=N)
data['Retinopathy'] = np.random.choice([0,1], size=N)
data['Neuropathy'] = np.random.choice([0,1], size=N)
data['CKD_Stage'] = np.random.choice([0,1,2,3,4,5], size=N)
data['Medication_Adherence_Rate'] = np.round(np.random.uniform(0,100, N),1)

# -------------------------------
# Multi-targets for all diseases
# -------------------------------
data['Target_Heart'] = np.where(data['Primary_Diagnosis']=='Heart Disease',1,0)
data['Target_Breast'] = np.where(data['Primary_Diagnosis']=='Breast Cancer',1,0)
data['Target_Lung'] = np.where(data['Primary_Diagnosis']=='Lung Cancer',1,0)
data['Target_Diabetes'] = np.where(data['Primary_Diagnosis']=='Diabetes',1,0)

# -------------------------------
# Save CSV
# -------------------------------
data.to_csv('Healthcare_Data_Multi_Disease.csv', index=False)

print("✅ Full 100k multi-disease dataset created!")
print("Shape:", data.shape)
print("\nPrimary Diagnosis Distribution:\n", data['Primary_Diagnosis'].value_counts())
print("\nSample targets distribution:\n", data[['Target_Heart','Target_Breast','Target_Lung','Target_Diabetes']].sum())
```
